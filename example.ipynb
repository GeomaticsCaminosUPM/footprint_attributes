```python
region = 'costa_rica'
base_path = 'C:\Topo\ProyectoExpoSismica\ProyectoPiloto\CostaRica'
footprints_path = base_path + f"\huella\huella_{region}.shp"
```

```python
footprints_gdf = gpd.read_file(footprints_path)
footprints_gdf = footprints_gdf.explode()
#footprints_gdf = footprints_gdf.loc[footprints_gdf['gid'].isin([1817,1811]),:]
footprints_gdf
```

```python
footprints_gdf = footprint_attributes.calc_forces(footprints_gdf,height_column=None,buffer=0.2) #20 cm of buffer
footprints_gdf
```

```python
footprints_gdf = footprint_attributes.relative_position(footprints_gdf,min_momentum=0.1389,min_confinement=1,min_angle=0.78,min_force=0.1666)
footprints_gdf
```

```python
footprints_gdf = footprint_attributes.calc_shape_irregularity(footprints_gdf)
footprints_gdf
```

```python
footprints_gdf = footprint_attributes.calc_polsby_popper(footprints_gdf)
footprints_gdf['shape_irregularity_pp'] = 1-((1-footprints_gdf['shape_irregularity']) * (footprints_gdf['polsby_popper_hull']))
footprints_gdf
```
